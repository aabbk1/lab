```python
def create_server(name, wt):
  return {"name": name, "weight": wt, "Rem": wt}

def get_ns(servers, ind):
  while True:
    server = servers[ind[0]]
    if server["Rem"]>0:
      server["Rem"] -= 1
      return server
    else:
      if all(s["Rem"] == 0 for s in servers):
        for s in servers:
          s["Rem"] = s["weight"]
      ind[0] = (ind[0]+1) % len(servers)

if __name__ == "__main__":
  servers = []
  n = int(input("no of server : "))
  for i in range(1, n+1):
    name = input("Name : ")
    wt = int(input("weight : "))
    servers.append(create_server(name, wt))

  num_load = int(input("no of load : "))
  curr_ind = [0]

  for i in range(1, num_load+1):
    saar = get_ns(servers, curr_ind)
    print(f'Load {i} assigned to server {saar["name"]}')
```
